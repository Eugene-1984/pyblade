```python
import numpy as np

import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
mpl.rcParams['figure.figsize'] = (14, 6)
```

```python
t = np.arange(64) * .01

```

```python
period = .22
length = .16

pulse = np.ones_like(t)  # array of 1's of the same shape as `t`
pulse[(t % period) > length] = 0  # generate pulse waveform with a given period and pulse length

plt.plot(t, pulse, marker='.')
```

```python
from scipy import signal

t = np.linspace(0, 1, 500)

plt.plot(t, signal.sawtooth(2 * np.pi * 5 * t))
```

```python
plt.plot(t, signal.square(2 * np.pi * 5 * t))
```
